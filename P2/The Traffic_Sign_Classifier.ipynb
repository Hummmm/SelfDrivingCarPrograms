````python
import csv

label_name_map = dict()

with open('signnames.csv') as f:
    reader = csv.reader(f)
    next(reader) # Skip the header row
    label_name_map = { int(rows[0]) : rows[1] for rows in reader }

def visualize(X, y):
    """
    Visualize images from the given
    dataset and their corresponding labels
    """
    columns = 5
    rows = len(X) / columns
    fig = plt.figure(figsize=(15,15))
    for i in range(len(X)):
        img = X[i]
        fig.add_subplot(rows,columns,i + 1)
        plt.title(label_name_map[y[i]])
        plt.imshow(img)

def visualize_random(X, y, num=5):
    """
    Visualize *num* ramdom images from the given
    sample and their corresponding labels
    """
    columns = 5
    rows = num / columns
    fig = plt.figure(figsize=(15,15))
    for i in range(num):
        idx = random.randint(0, len(X))
        img = X[idx].squeeze()
        fig.add_subplot(rows,columns,i + 1)
        plt.title(label_name_map[y[idx]])
        plt.imshow(img, cmap="gray")
````

````python
import tensorflow as tf
import cv2 as cv

kernel = np.ones((3,3), np.uint8)

### Augment Data by applying common OpenCV transforms
def gaussian_blur_transform(img):
    """
    http://docs.opencv.org/3.1.0/d4/d13/tutorial_py_filtering.html
    """
    return cv.GaussianBlur(img, (3, 3), 0)

def rescale_transform(img):
    """
    http://docs.opencv.org/3.1.0/da/d6e/tutorial_py_geometric_transformations.html
    """
    rows,cols = img.shape[:2]
    scale = random.randint(90, 110) * 0.01
    small = cv.resize(img, None, fx=scale, fy=scale, interpolation = cv.INTER_CUBIC)
    return cv.resize(small, (rows, cols))

def translate_transform(img):
    """
    http://docs.opencv.org/3.1.0/da/d6e/tutorial_py_geometric_transformations.html
    """
    rows,cols = img.shape[:2]
    scale = random.randint(-3, 3)
    M = np.float32([[1,0,scale],[0,1,scale]])
    return cv.warpAffine(img,M,(cols,rows))

def brg_2_gray_transform(img):
    """
    http://docs.opencv.org/3.1.0/df/d9d/tutorial_py_colorspaces.html
    """
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    # Convert to 3 channels so that it is easier to process along
    # with other images
    return cv.cvtColor(gray, cv.COLOR_GRAY2BGR)

def opening_morph_transform(img):
    """
    http://docs.opencv.org/3.1.0/d9/d61/tutorial_py_morphological_ops.html
    """
    return cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

def closing_morph_transform(img):
    """
    http://docs.opencv.org/3.1.0/d9/d61/tutorial_py_morphological_ops.html
    """
    return cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

def rotate_transform_clock(img):
    """
    http://docs.opencv.org/3.1.0/da/d6e/tutorial_py_geometric_transformations.html
    """
    rows,cols = img.shape[:2]
    rot = random.randint(-15, 1)
    M = cv.getRotationMatrix2D((cols/2,rows/2),rot,1)
    return cv.warpAffine(img,M,(cols,rows))

def rotate_transform_anticlock(img):
    """
    http://docs.opencv.org/3.1.0/da/d6e/tutorial_py_geometric_transformations.html
    """
    rows,cols = img.shape[:2]
    rot = random.randint(1, 15)
    M = cv.getRotationMatrix2D((cols/2,rows/2),rot,1)
    return cv.warpAffine(img,M,(cols,rows))

transform_func_list = [
    gaussian_blur_transform,
    rescale_transform,
    translate_transform,
    opening_morph_transform,
    closing_morph_transform,
    rotate_transform_clock,
    rotate_transform_anticlock]

def apply_random_transform(img, times = 1):
    """
    Randomly apply *times* number of transform 
    from a list of transforms
    :param img: Image to apply transform on
    :param times: Number of random transforms to apply
    :return: Transformed image data
    """
    img_new = []
    for i in range(times):
        index = random.randint(0, len(transform_func_list) - 1)
        random_tranform_function = transform_func_list[index]
        img_new.append(random_tranform_function(img))
    return img_new
````

````python
def augment_data(X, y, mean, stddev, counts, times=7):
    """
    Augment classes with low number of examples by
    applying a random transform from a list of transform
    and returning the transformed dataset
    :param X_train: The training features
    :param y_train: The training labels
    :return: Normalized image data
    """
    X_augment = []
    y_augment = []
    for idx in range(len(X)):
        if counts[y[idx]] < (mean + stddev):
            transformed_imgs = apply_random_transform(X[idx], times)
            X_augment.extend(transformed_imgs)
            y_augment.extend(times * [y[idx]])
            counts[y[idx]] += times
    return X_augment, y_augment
````

````python
## Normalize all data
def histogram_eq_transform(img):
    """
    http://docs.opencv.org/3.1.0/d5/daf/tutorial_py_histogram_equalization.html
    """
    rows,cols = img.shape[:2]
    gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
    equalized = cv.equalizeHist(gray)
    return equalized.reshape(rows,cols,1)

def normalize(image_data):
    """
    Normalize the image data using (pixel - 128)/ 128
    :param image_data: The image data to be normalized
    :return: Normalized image data
    """
    return ((image_data - 128.) / 128.)

def preprocess(image_date):
    hist_equalizer = np.asarray([histogram_eq_transform(x) for x in image_date])
    return normalize(hist_equalizer)

X_train = preprocess(X_train)
X_valid = preprocess(X_valid)
X_test = preprocess(X_test)

visualize_random(X_train, y_train)
````

````python
from sklearn.utils import shuffle

X_train, y_train = shuffle(X_train, y_train)
````

````python
import tensorflow as tf

EPOCHS        = 20
BATCH_SIZE    = 128
DROPOUT       = 0.75
MU            = 0
SIGMA         = 0.1
LEARNING_RATE = 0.001
````

````python
def conv2d(x, W, b, name, strides=1, padding='VALID'):
    x = tf.nn.conv2d(x, W, strides=[1, strides, strides, 1], padding=padding, name=name)
    x = tf.nn.bias_add(x, b)
    return tf.nn.relu(x)

def maxpool2d(x, k=2, padding='VALID'):
    return tf.nn.max_pool(x, ksize=[1, k, k, 1], strides=[1, k, k, 1], padding=padding)
````

````python
from tensorflow.contrib.layers import flatten

def NNModel(x, dropout):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    
    # Layer 1a: Convolutional + Activation. Input = 32x32x1. Output = 30x30x6.
    conv1a_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 1, 6), mean = MU, stddev = SIGMA))
    conv1a_b = tf.Variable(tf.zeros(6))
    conv1a = conv2d(x, conv1a_W, conv1a_b, 'conv1a')

    # Layer 1b: Convolutional + Activation. Input = 30x30x6. Output = 28x28x12.
    conv1b_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 6, 12), mean = MU, stddev = SIGMA))
    conv1b_b = tf.Variable(tf.zeros(12))
    conv1b = conv2d(conv1a, conv1b_W, conv1b_b, 'conv1b')

    # Pooling. Input = 28x28x12. Output = 14x14x12.
    conv1b = maxpool2d(conv1b)

    # Layer 2a: Convolutional + Activation. Output = 12x12x18.
    conv2a_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 12, 18), mean = MU, stddev = SIGMA))
    conv2a_b = tf.Variable(tf.zeros(18))
    conv2a   = conv2d(conv1b, conv2a_W, conv2a_b, 'conv2a')

    # Layer 2b: Convolutional + Activation. Output = 10x10x24.
    conv2b_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 18, 24), mean = MU, stddev = SIGMA))
    conv2b_b = tf.Variable(tf.zeros(24))
    conv2b   = conv2d(conv2a, conv2b_W, conv2b_b, 'conv1b')

    # Pooling. Input = 10x10x16. Output = 5x5x24.
    conv2b = maxpool2d(conv2b)

    # Flatten. Input = 5x5x24. Output = 600.
    fc0   = flatten(conv2b)
    
    # Layer 3: Fully Connected. Input = 600. Output = 120.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(600, 120), mean = MU, stddev = SIGMA))
    fc1_b = tf.Variable(tf.zeros(120))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    
    # Activation.
    fc1    = tf.nn.relu(fc1)

    # Dropout.
    fc1    = tf.nn.dropout(fc1, dropout)

    # Layer 4: Fully Connected. Input = 120. Output = 84.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(120, 84), mean = MU, stddev = SIGMA))
    fc2_b  = tf.Variable(tf.zeros(84))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # Activation.
    fc2    = tf.nn.relu(fc2)

    # Dropout.
    fc2    = tf.nn.dropout(fc2, dropout)

    # ayer 5: Fully Connected. Input = 84. Output = 43.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(84, 43), mean = MU, stddev = SIGMA))
    fc3_b  = tf.Variable(tf.zeros(43))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return logits
````

````python
x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, 43)
dropout = tf.placeholder(tf.float32)
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
````

````python
### TRAINING

logits = NNModel(x, dropout)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = LEARNING_RATE)
training_operation = optimizer.minimize(loss_operation)
````

````python
### EVALUATION
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, dropout: DROPOUT})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
### TRAIN THE MODEL
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, dropout: DROPOUT})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()

    saver.save(sess, './lenet')
    print("Model saved")
````

````python
### EVALUATE THE MODEL
with tf.Session() as sess:
    saver.restore(sess, tf.train.latest_checkpoint('.'))

    test_accuracy = evaluate(X_test, y_test)
    print("Test Accuracy = {:.3f}".format(test_accuracy))
````

````python
def top_five_viz(probs, labels):
    """
    https://matplotlib.org/examples/lines_bars_and_markers/barh_demo.html
    """
    for idx in range(len(labels)):
        traffic_sign_text = [label_name_map[x] for x in labels[idx]]
        y_pos = np.arange(len(traffic_sign_text))
        prob = probs[idx]

        fig, ax = plt.subplots()
        fig.set_figheight(2)
        fig.set_figwidth(2)
        ax.barh(y_pos, prob, align='center', ecolor='black')
        ax.set_yticks(y_pos)
        ax.set_yticklabels(traffic_sign_text)
        ax.invert_yaxis()  # labels read top-to-bottom
        ax.set_xlabel('Probability')
        ax.set_title('Unseen Image: {}'.format(idx + 1))
        plt.show()

top_five_viz(probs, labels)
````
